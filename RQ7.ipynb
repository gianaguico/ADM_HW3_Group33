## Algorithms

### Selectionsort

**Selection sort** is a sorting algorithm that selects the smallest element from an unsorted list in each iteration and places that element at the beginning of the unsorted list.

```python
def selection_sort(arr):
  #We give the indexes of the columns  
    first_name = 0
    last_name = 1
    mean = 2
    
    #Iteration over the lenght of the array
    for i in range(len(arr)):
      #To set the min index
        min_ind = i
        
        #This works over the rest of the array oder the index
        for element in range(i+1, len(arr)):
          #To select the minimum element in each iteartion
          if arr[element][mean] > arr[min_ind][mean] or arr[element][mean] == arr[min_ind][mean] and arr[element][first_name] < arr[min_ind][first_name]:
            min_ind = element  

        arr[i], arr[min_ind] = arr[min_ind], arr[i].copy()    #To set the index at the correct position and swaps the elements to sort the array
                                                              #We create a copy of the existing array in order to not modify it
        
    return arr     #This returns the sorted array
```

```python
%%time
SelectionSort_arr = selection_sort(df_arr)
print(SelectionSort_arr)
```

```python
#creation of the txt file
SelectionSort_df = pd.DataFrame(SelectionSort_arr)
SelectionSort_df.to_csv(r'RankingList_SS.txt', header=None, index=None, sep=' ', mode='a')
```

#### **Time Complexity**
In the selection sort algorithm the principle is to reduce by one the number of comparisons at each iteration.  
at the first step, the number of comparisons is n-1. At the second the algorithm has to scan a shorter array, and so on, until the array has a length of one. 

$\ (n-1)+(n-2)+...+1 $= $\sum\limits_{i=0}^{n-1} i$

This is an arithmetic progession:

$\sum_{i=0}^{n-1} i$ = $\sum_{j=i+1}^{n} j$

and converges to $\frac{n(n-1)}{2}$

Hence, our sorting algorithm has polynomial complexity.

**$\Theta(n)$ = $\ n^2$**




### Mergesort

**Merge Sort** is a sorting algorithm that is based on the principle of Divide and Conquer (Divide et Impera), which divides the problem into subproblems. When the solution to each subproblem is ready, it 'combines' the results from the subproblems to solve the main problem.

```python

def merge_sort(arr):
  #Give the indexes of the columns
    first_name = 0
    last_name = 1
    mean = 2
       
    if len(arr) > 1:
        mid = len(arr)//2         #This is the point where the array is divided in half
        L = arr[:mid].copy()      #We define the Left and Right halves, creating a copy of the existing array in order to not modify it
        R = arr[mid:].copy()
        merge_sort(L)             #Sort the two halves
        merge_sort(R)
        
        i = j = k = 0
        
        #Until we reach either end of either L or R, this picks the largest elements among
        #L and R and places them in the correct position into the array

        while (i < len(L) and j < len(R)):
          if L[i][mean] < R[j][mean] or L[i][mean] == R[j][mean] and L[i][first_name] > R[j][first_name]:
            arr[k] = R[j]
            j += 1
          else: 
            arr[k] = L[i]
            i += 1 
          k += 1
       #When we run out of elements in either L or R, this
       #picks up the remaining elements and put into the array

        if i >= len(L):
          arr[k:] = R[j:]
        else:
          arr[k:] = L[i:]
        
    return arr                #To return the sorted array
```

```python
%%time
mergeSort_arr = merge_sort(df_arr)
print(mergeSort_arr)
```

```python
#creating the txt file
mergeSort_df = pd.DataFrame(mergeSort_arr)
mergeSort_df.to_csv(r'RankingList_MS.txt', header=None, index=None, sep=' ', mode='a')
```

### **Time Complexity**
if n is the length of the array the function T(n) of our algorithm is given by the sum of: 
- $\ T(n/2)$ for the left array

- $\ T(n/2)$ for the right array

- $\ T(n)$ 

The algorithm is indeed applied to the two arrays of half the length of the original one and requires the merge operation (n iterations).

Hence: 

**$\Theta(n)$ = $\ nlog(n)$**




### **Quicksort**

**Quicksort** is a sorting algorithm based on the divide and conquer approach.

As a first thing, the array is divided into subarrays by selecting a pivot element, which is an element selected from the array.

Then, while dividing the array, the pivot element should be positioned in a way that elements smaller than pivot are kept on the left side, and elements greater than pivot are on the right side.

Then, left and right subarrays are also divided using the same approach. This process continues until each subarray contains a single element, which is considered already sorted.

Finally, elements are combined and the sorted array is formed.

```python
def quick_sort(arr):
  #Give the indexes of the two columns
  first_name = 0
  last_name = 1
  mean = 2
    
  #Base case
  if len(arr) < 2:
    return arr
    
    current_position = 0            #Position of the partitioning element

    for i in range(1, len(arr)):     #Partitioning loop
         if arr[i][mean] <= arr[0][mean] or arr[i][mean] == arr[0][mean] and arr[i][first_name] > arr[0][first_name]:   #Requested condition over the mean and first name
              current_position += 1
              temp = arr[i]
              arr[i] = arr[current_position]
              arr[current_position] = temp

    temp = arr[0]
    arr[0] = arr[current_position] 
    arr[current_position] = temp              #Brings pivot to the appropriate position
    
    left = quick_sort(arr[0:current_position].copy())                #Sorts the elements to the left of pivot
    right = quick_sort(arr[current_position+1:len(arr)].copy())      #Sorts the elements to the right of pivot
    #We create a copy of the existing array in order to not modify it

    arr = left + [arr[current_position]] + right                      #Merging everything together
    
    return arr                 #To return the sorted array
```

```python
%%time
quickSort_arr = quick_sort(df_arr)
print(quickSort_arr)
```

```python
#creating the txt file
quickSort_df = pd.DataFrame(quickSort_arr)
quickSort_df.to_csv(r'RankingList_QS.txt', header=None, index=None, sep=' ', mode='a')
```

### **Time Complexity**

In the worst case(the pivot element is either the greatest or the smallest element in the array) the partition process is "extremely" unbalanced (one sub-array has n-1 elemtent and the other 0). If this unbalanced partitioning arises at each recursive call: 
$\ T(n) = T(n-1) +c*n$

Hence 
$\Theta(n)$ = $\ n^2$
