## Data import e calcolo della media dei voti

```python
df_grades = pd.read_csv('Algoritmo_prova.txt', header= None, skiprows=1, delim_whitespace=True)
print(df_grades)
```

```python
num_sum = df_grades.sum(numeric_only=True, axis=1)
#num_sum
```

```python
df_grades['mean'] = num_sum/((df_grades.shape[1]-2))
df_grades['mean'] = np.round(df_grades['mean'], decimals=2)
print(df_grades)
```
